```python
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
attributes = pd.read_csv("../raw_data/player_attributes.csv")
```

```python
attributes
```

```python
attributes_nokeeper = attributes.query("player_position_primary != 'Keeper'")
```

```python
attributes.drop("player_name", axis=1, inplace=True)
attributes.drop("team_id", axis=1, inplace=True)
attributes.drop("team_color", axis=1, inplace=True)
attributes.drop("team", axis=1, inplace=True)
attributes.drop("player_position_primary", axis=1, inplace=True)
attributes.drop("player_position_nonprimary", axis=1, inplace=True)
attributes.drop("player_preferred_foot", axis=1, inplace=True)
attributes.drop("player_age", axis=1, inplace=True)
attributes.drop("player_country", axis=1, inplace=True)
attributes.drop("player_shirt", axis=1, inplace=True)
attributes.drop("player_market_value", axis=1, inplace=True)
attributes.drop("player_height", axis=1, inplace=True)
```

```python
attributes
```

```python
ratings = pd.read_csv("../raw_data/samson/fotmob-premier-league-players-matches.csv")
```

```python
ratings = ratings[['player_id','match_player_rating']]
```

```python
ratings = ratings.merge(attributes, how='left', on='player_id')
```

```python
ratings = ratings[ratings['match_player_rating']!=0]
```

```python
ratings.to_csv("../raw_data/attributes_ratings_removed_zero_removed_keeper.csv")
```

```python
X = ratings.drop(columns=['player_id','match_player_rating'])
```

```python
y = ratings['match_player_rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LinearRegression()
```

```python
attributes[attributes.isna().any(axis=1)]
```

```python
X_train[X_train.isna().any(axis=1)]
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
