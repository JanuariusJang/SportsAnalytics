```python
resultsOnly_data=results_data[["Season", "HomeTeam", "AwayTeam", "FTHG", "FTAG"]]
```

```python
def label_WDL (row):
    if row['FTHG']>row['FTAG']:
        return 'W'
    if row['FTHG']<row['FTAG']:
        return 'L'
    if row['FTHG']==row['FTAG']:
        return 'D'
```

```python
resultsOnly_data['WDL']=resultsOnly_data.apply(lambda row: label_WDL(row), axis=1)
```

```python
def winning_team (row):
    if row['WDL']=='W':
        return row['HomeTeam']
    if row['WDL']=='L':
        return row['AwayTeam']
    if row['WDL']=='D':
        return "NA"
```

```python
resultsOnly_data['WinningTeam']=resultsOnly_data.apply(lambda row: winning_team(row), axis=1)
```

```python
resultsOnly_data.drop('WinningTeam', axis=1, inplace=True)
resultsOnly_data.drop('Season', axis=1, inplace=True)
resultsOnly_data.drop('FTHG', axis=1, inplace=True)
resultsOnly_data.drop('FTAG', axis=1, inplace=True)
```

```python
resultsOnly_data.replace("Nott\'m Forest", "Nott Forest", inplace=True)
```

```python
for team in team_list:
    query = f"HomeTeam == '{team}' or AwayTeam == '{team}'"
    temp = resultsOnly_data.query(str(query))
    temp_index = list(temp.index)
    form = ['NA','NA','NA','NA','NA']
    for i in range(5, len(temp_index)):
        form.append(temp.to_numpy()[i-5][2] + temp.to_numpy()[i-4][2] + temp.to_numpy()[i-3][2] + temp.to_numpy()[i-2][2] + temp.to_numpy()[i-1][2])
    form_numerical = []
    for item in form:
        num = 0
        for letter in item:
            if letter == 'W':
                num += 2
            elif letter == 'D':
                num += 1
        form_numerical.append(num)
    temp['Form Numerical'] = form_numerical
    temp = temp.iloc[5:]
    temp.to_csv(f'../raw_data/test/{team}.csv')
```
