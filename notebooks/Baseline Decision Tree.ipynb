```python
import pandas as pd
import numpy as np


from sklearn import tree
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("../raw_data/final_form.csv")
```

```python
data
```

```python
X = data[['HomeTeam','AwayTeam', 'Home Form', 'Away Form']]
```

```python
ohe = OneHotEncoder(sparse=False)
```

```python
enc_data = pd.DataFrame(ohe.fit_transform(X[['HomeTeam','AwayTeam']]))
```

```python
enc_data
```

```python
ohe.categories_
```

```python
ohe.get_feature_names_out()
enc_data.columns = ohe.get_feature_names_out()
```

```python
enc_data
```

```python
X.drop('HomeTeam', axis=1, inplace=True)
X.drop('AwayTeam', axis=1, inplace=True)
```

```python
X = enc_data.merge(X, left_index=True, right_index=True)
```

```python
X
```

```python
y = data['WDL']
```

```python
y
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(X,y)
```

```python
test = 1

res = clf.predict_proba(X.loc[test].to_numpy().reshape(1,-1))[0]

print(res)
print(y.loc[test])
```

```python
X.loc[test].to_numpy().reshape(1,-1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test.to_numpy()[0]
```

```python
def accuracy_score(y_pred, y_test):
    correct = 0
    y_test_array = y_test.to_numpy()
    
    for i in range(0, len(y_pred)-1):
        if y_pred[i] == y_test_array[i]:
            correct += 1
        
    accuracy = correct/len(y_pred) * 100
    
    return accuracy
```

```python
accuracy = accuracy_score(y_pred, y_test)
```
