```python
import numpy as np
import pandas as pd
results_data = pd.read_csv('/Users/januariusjang/code/JanuariusJang/sportsanalytics/data/external/Past_results_1993_2021.csv', encoding='windows-1252')
results_data.head()
```

```python
resultsOnly_data=results_data[["Season", "HomeTeam", "AwayTeam", "FTHG", "FTAG"]]
```

```python
resultsOnly_data.tail()
```

```python
def label_WDL (row):
    if row['FTHG']>row['FTAG']:
        return 'W'
    if row['FTHG']<row['FTAG']:
        return 'L'
    if row['FTHG']==row['FTAG']:
        return 'D'
```

```python
resultsOnly_data['WDL']=resultsOnly_data.apply(lambda row: label_WDL(row), axis=1)
```

```python
resultsOnly_data.head()
```

```python
def winning_team (row):
    if row['WDL']=='W':
        return row['HomeTeam']
    if row['WDL']=='L':
        return row['AwayTeam']
    if row['WDL']=='D':
        return "NA"
```

```python
resultsOnly_data['WinningTeam']=resultsOnly_data.apply(lambda row: winning_team(row), axis=1)
```

```python
resultsOnly_data.head()
```

```python
resultsOnly_data['WinningTeam'].value_counts()
```

```python
Club_WinHistory=resultsOnly_data['WinningTeam'].value_counts()
```

```python
Club_WinHistory=Club_WinHistory.tail(-1)
```
